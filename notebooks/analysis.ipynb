```python
import sqlite3
import pandas as pd
from tabulate import tabulate 

conn = sqlite3.connect('../database/ubereats_project.db')
cursor=conn.cursor()

print("system check: Connection Active.")


cursor.execute("select name from sqlite_master where type='table';")
print(f"Table Ready: {cursor.fetchall()} ")
```

```python
df_view=pd.read_sql('select * from restaurants LIMIT 10', conn)
df_view
```

```python
# 1. Which Bangalore locations have the highest average restaurant ratings?

query = """
SELECT location, round(AVG(rate_norm), 2) as avg_rating
FROM restaurants
GROUP BY location
ORDER BY avg_rating DESC
LIMIT 10;
"""

cursor.execute(query)
results=cursor.fetchall()
print(tabulate(results, headers=["Locations", "Avg Rating"], tablefmt="fancy_grid"))
```
